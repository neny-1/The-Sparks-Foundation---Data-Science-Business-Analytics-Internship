# Import libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

# Read the data

```python
data_ulr = 'https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
data = pd.read_csv(data_ulr)
```

```python
data.describe()
```

```python
data.head()
```

```python
X=data['Hours']
y=data['Scores']
```

```python
X = X.values.reshape(-1,1)
y = y.values.reshape(-1,1)
```

# data visualization

```python
plt.scatter(X,y,color = 'blue',marker="v")
plt.style.use('seaborn')
plt.xlabel("Hours",fontsize = 20)
plt.ylabel("Scores",fontsize = 20)
plt.title("Hours vs Scores",fontsize=20,pad=20)
plt.legend()
```

# split the data to test and train

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25)
```

#  bold textBuild  linear regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# visalize the model 
line = model.coef_*X+model.intercept_
plt.plot(X,line,label="Linear Regression",color="b")
plt.scatter(X,y,label="Actual Test Data",color='g',marker="v")
plt.xlabel("Hours",fontsize = 20)
plt.ylabel("Scores",fontsize = 20)
plt.title("Hours vs Scores",fontsize=20,pad=20)
plt.show()
```

## Prediction

```python
y_pred = model.predict(X_test)
```

## Evaluate the model

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```

```python
from sklearn import metrics  
print('Mean squared Error:', 
      metrics.mean_squared_error(y_test, y_pred)) 
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

## predict if the student studies 9.5 hrs/day

```python
student_hours = np.array([[9.25]])
student_score= model.predict(student_hours)
print('If the student study',student_hours,'hours the student score will be',student_score)
```
